# Subsetting Data

```python
# Subsetting columns
df = df[['title', 'description', 'publishTime',  'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
```

```python
df.shape
```

```python
# Subsetting rows usinng loc
df_subset1 = df.loc[df["viewCount"] > 1000000]
```

```python
# Subsetting using query
df_subset2 = df.query("viewCount > 1000000")
```

```python
df = df.loc[~df['viewCount'].isna()]  # with ~df['viewCount...'] we do the opposite, the Na value become false and the other true
df = df.loc[~df['likeCount'].isna()]
```

# Casting dtypes

```python
df['viewCount']= df['viewCount'].astype('int')
df['likeCount']= df['likeCount'].astype('int')
```

```python
df.dtypes
```

```python
df['publishTime']= pd.to_datetime(df['publishTime'])
```

```python
df['likeCount'] = pd.to_numeric(df['likeCount'].astype('str'))
```

# Creating new column

```python
df['like_to_view_raito'] = df['likeCount'] / df['viewCount']
df.head()
```

# adding new row


```python
df_to_append= df.tail(1)
```

```python
df_concat = pd.concat([df,df_to_append])
```
